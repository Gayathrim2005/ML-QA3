```python
# Training Data
emails = [
    ("free offer", "Spam"),
    ("free", "Spam"),
    ("offer", "Spam"),
    ("hello friend", "Not Spam"),
    ("good morning", "Not Spam")
]

# Count classes
spam_count = 0
notspam_count = 0

for text, label in emails:
    if label == "Spam":
        spam_count += 1
    else:
        notspam_count += 1

total = len(emails)

# Prior Probabilities
p_spam = spam_count / total
p_notspam = notspam_count / total

# Likelihoods
free_spam = 2/3
offer_spam = 2/3

free_notspam = 0/2
offer_notspam = 0/2

# Naive Bayes Calculation
spam_score = p_spam * free_spam * offer_spam
notspam_score = p_notspam * free_notspam * offer_notspam

print("P(Spam) =", p_spam)
print("P(Not Spam) =", p_notspam)

print("\nSpam Score =", spam_score)
print("Not Spam Score =", notspam_score)

if spam_score > notspam_score:
    print("\nPrediction: Spam")
else:
    print("\nPrediction: Not Spam")
```
